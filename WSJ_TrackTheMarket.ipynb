```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# YTD Chg
df=pd.read_csv('https://wsj-scoreboard.s3.amazonaws.com/WSJTrackMarkets_latest.csv', index_col=['Name'], parse_dates=['TradeDate']).sort_values(['YTD % Chg'], ascending=False)
```

```python
df.head()
```

```python
df.tail()
```

```python
f, ax = plt.subplots(figsize=(12, 24))
sns.set_color_codes("pastel")
sns.barplot(x=df['YTD % Chg'], y=df.index, data=df,label="Total", color="orange")
```

```python
df_chg_WTD = df.sort_values(['WTD % Chg'], ascending=False)
df_chg_QTD = df.sort_values(['QTD % Chg'], ascending=False)
df_chg_YTD = df.sort_values(['YTD % Chg'], ascending=False)
```

### 1. WTD Chg

```python
WTD_head = df_chg_WTD.head()
WTD_tail = df_chg_WTD.tail()
```

```python
WTD_head
```

```python
WTD_tail
```

```python
y_h_WTD = WTD_head['WTD % Chg'].sort_values()
x_h_WTD = WTD_head['WTD % Chg'].sort_values().index
```

```python
x_t_WTD = WTD_tail.index
y_t_WTD = WTD_tail['WTD % Chg']
```

```python
plt.figure(figsize=(8,4))

plt.subplot(211)
plt.barh(x_h_WTD, y_h_WTD, align='center', alpha=0.7, color='Red')
plt.yticks(x_h_WTD)
plt.title('Weekly Hot 5 & Cool 5')

plt.subplot(212)
plt.barh(x_t_WTD, y_t_WTD, align='center', alpha=0.7, color='Blue')
plt.yticks(x_t_WTD)
plt.xlabel('Change(%)')


plt.show()
```

### 2. QTD Chg

```python
QTD_head = df_chg_QTD.head()
QTD_tail = df_chg_QTD.tail()
```

```python
y_h_QTD = QTD_head['QTD % Chg'].sort_values()
x_h_QTD = QTD_head['QTD % Chg'].sort_values().index
```

```python
x_t_QTD = QTD_tail.index
y_t_QTD = QTD_tail['QTD % Chg']
```
